Rescaling a feature

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("Finance_data.csv")

# Automatically detect numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", numeric_cols)

scaler = MinMaxScaler()

df_scaled = df.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print(df_scaled.head())
```

Standardizing a Feature

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler

df = pd.read_csv("Finance_data.csv")

# detect numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", num_cols)

# StandardScaler
std_scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[num_cols] = std_scaler.fit_transform(df[num_cols])

# RobustScaler
rob_scaler = RobustScaler()
df_robust_scaled = df.copy()
df_robust_scaled[num_cols] = rob_scaler.fit_transform(df[num_cols])

print("\nStandard Scaled:")
print(df_standard_scaled.head())

print("\nRobust Scaled:")
print(df_robust_scaled.head())
```

Normalizing Observations


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Normalizer

df = pd.read_csv('Finance_data.csv')

feature_cols = ['Mutual_Funds', 'Equity_Market', 'Debentures',
                'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
features = df[feature_cols].values

normalizer_l2 = Normalizer(norm="l2")
l2_normalized = normalizer_l2.transform(features)
l2_normalized_df = pd.DataFrame(l2_normalized, columns=[f'{col}_L2_Norm' for col in feature_cols])

print("--- L2 Normalized Features (First 5 rows) ---")
print(l2_normalized_df.head())

normalizer_l1 = Normalizer(norm="l1")
l1_normalized = normalizer_l1.transform(features)
l1_normalized_df = pd.DataFrame(l1_normalized, columns=[f'{col}_L1_Norm' for col in feature_cols])

print("\n--- L1 Normalized Features (First 5 rows) ---")
print(l1_normalized_df.head())

print("\nVerification (Sum of the first L1 Normalized row):", l1_normalized_df.iloc[0].sum())

# df_final = pd.concat([df, l2_normalized_df, l1_normalized_df], axis=1)
```

Grouping Observations Using Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('Finance_data.csv')

clustering_features = [
    'Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds',
    'Fixed_Deposits', 'PPF', 'Gold', 'age'
]

features = df[clustering_features].copy()

features.fillna(features.mean(), inplace=True)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

scaled_df = pd.DataFrame(scaled_features, columns=clustering_features)

clusterer = KMeans(n_clusters=3, random_state=0, n_init='auto')

clusterer.fit(scaled_df)

df['Cluster_Group'] = clusterer.predict(scaled_df)

print("--- Data after K-Means Clustering (First 5 Rows) ---")
print(df[['age', 'Avenue', 'Expect', 'Cluster_Group']].head())

print("\n--- Cluster Sizes ---")
print(df['Cluster_Group'].value_counts().sort_index())

cluster_avenue_preference = df.groupby('Cluster_Group')['Avenue'].value_counts(normalize=True).mul(100).rename('Percentage').reset_index()
print("\n--- Investment Avenue Preference by Cluster ---")
print(cluster_avenue_preference)
```

Imputing Missing Values

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

df = pd.read_csv("Finance_data.csv")
features = df.select_dtypes(include=['int64', 'float64']).values

scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

true_value = standardized_features[0, 0]
standardized_features[0, 0] = np.nan

mean_imputer = SimpleImputer(strategy="mean")
features_mean_imputed = mean_imputer.fit_transform(standardized_features)

print(true_value)
print(features_mean_imputed[0, 0])
```
