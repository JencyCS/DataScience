Encoding Nominal Categorical Feature

One-hot encode the feature using scikit-learn's LabelBinarizer

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelBinarizer

df = pd.read_csv("Finance_data.csv", header=None, engine="python")

# column 0 contains the items
feature = df[0].values

# LabelBinarizer needs 1D input
one_hot = LabelBinarizer()
encoded = one_hot.fit_transform(feature)

print(encoded[:10])
print(one_hot.classes_)
```

```python
# Reverse one-hot encoding
decoded = one_hot.inverse_transform(encoded)
print("\nDecoded (Reverse One-Hot):\n", decoded)
```

```python
# One hot encode with pandas
pd.get_dummies(df.iloc[:, 10])
```

```python
import pandas as pd

df = pd.read_csv("Finance_data.csv", header=None, sep="\t", engine="python")
print(df.head(10))
```

```python
transactions = df.apply(lambda row: row.dropna().tolist(), axis=1)
transactions = transactions.tolist()

print(transactions[:5])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

encoded = mlb.fit_transform(transactions)

print(encoded[:5])
print(len(mlb.classes_), "unique items")
print(mlb.classes_)
```

Handling Imbalanced Classes

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

iris = load_iris()

features = iris.data
target = iris. target

# remove first 40 observations
features = features [40:, : ]
target = target [40:]

# create binary target vector indicating if class 0
target = np.where((target == 0), 0, 1)
target
```

```python
# create weights
weights = {0: .9, 1: 0.1}

# Create random forest classifier with weights
RandomForestClassifier(class_weight=weights)

RandomForestClassifier(class_weight='balanced')

i_class0 = np.where(target == 0) [0]
i_class1 = np.where(target == 1) [0]

n_class0 = len(i_class0)
n_class1 = len(i_class1)

i_class1_downsampled = np.random.choice(i_class1, size=n_class0, replace=False)

np.hstack((target[i_class0], target[i_class1_downsampled]))

```

```python
np.vstack((features[i_class0,:], features[i_class1_downsampled, : ] ) ) [0:5]
```
