```python
# Load required libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
data = load_iris()

# Create DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add species column
df["species"] = pd.Categorical.from_codes(data.target, data.target_names)

# --- Your filters ---

# Condition 1: species is setosa
setosa = df['species'] == 'setosa'

# Condition 2: sepal length >= 5
long_sepal = df['sepal length (cm)'] >= 5

# Print top 2 setosa rows
print("Top 2 setosa rows:")
print(df[setosa].head(2))

print("\nSetosa with sepal length >= 5:")
print(df[setosa & long_sepal])
```

```python
import pandas as pd

# Sample DataFrame
data = {'Sex': ['female', 'male', 'female', 'male', 'female']}
df = pd.DataFrame(data)

# Replace 'female' with 'Woman' and 'male' with 'Man'
df['Sex'] = df['Sex'].replace(['female', 'male'], ['Woman', 'Man'])

# Display the modified DataFrame
print(df.head(5))
```

```python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
df = sns.load_dataset("titanic")

# Display the first few rows of the DataFrame
print(df.head())

# Renaming columns
df.rename(columns={'pclass': 'Passenger Class', 'sex': 'Gender'}, inplace=True)

# Check the data types
print(df.dtypes)

# Handle non-numeric values in the 'age' column
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Finding the Min, Max, Sum, Average, and Count for the 'age' column
print(f"Maximum Age: {df['age'].max()}")
print(f"Minimum Age: {df['age'].min()}")
print(f"Mean Age: {df['age'].mean()}")
print(f"Sum of Ages: {df['age'].sum()}")
print(f"Count of Ages: {df['age'].count()}")

```
